```python
import matplotlib.pyplot as plt
epoch=50
# 
file_path = f"Summaries/SL_module/visual_ep{epoch}_11240145/logs/loss.txt"  # 
epochs = []
losses = []
val_losses = []

with open(file_path, "r") as file:
    for line in file:
        if line.startswith("Epoch:"):
            parts = line.split(",")
            epoch = int(parts[0].split(":")[1].strip())
            loss = float(parts[1].split(":")[1].strip())
            val_loss = float(parts[2].split(":")[1].strip())
            epochs.append(epoch)
            losses.append(loss)
            val_losses.append(val_loss)

# 
plt.figure(figsize=(10, 6))
plt.plot(epochs, losses, color='red', label='Loss')
plt.plot(epochs, val_losses, color='blue', label='Validation Loss')

# 
plt.title(f"Loss and Validation Loss Over {epoch} Epochs", fontsize=16)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)

# 
plt.legend()

# 
plt.grid(True)
file_path = f"Summaries/SL_module/visual_ep{epoch}_11240145/logs/Epoch_{epoch}.png"
plt.savefig(file_path)
# 
plt.show()
```
